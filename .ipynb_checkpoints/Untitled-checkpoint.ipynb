```python
from sklearn.model_selection import train_test_split 
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pylab as plt
import random
import math
import os
```

```python
df = pd.read_csv("../data/heart.csv")
df.head()
```

```python
is_categorical = ["Sex","ChestPainType","FastingBS","RestingECG","ExerciseAngina","ST_Slope"]
for col in is_categorical: # this serves to show the range of each categorical feature while looking for missing datapoints.
    print(f'The unique values in {col} are: \n {df[col].unique()}')
    pd.value_counts(df[col]).plot.bar()
    plt.ylabel('count')
    plt.xlabel(col)
    plt.title(col + ' Bar Plot')
    plt.savefig(f"../figures/{col}_bar.png",bbox_inches="tight")
    plt.show()
```

```python

for col in is_categorical: # this serves to show the range of each categorical feature while looking for missing datapoints.
    count_matrix = df.groupby([col, 'HeartDisease']).size().unstack()
    count_matrix_norm = count_matrix.div(count_matrix.sum(axis=1),axis=0)
    count_matrix_norm.plot(kind='bar', stacked=True)
    plt.ylabel('fraction of people with Heart Disease')
    plt.legend(["Heart Disease","Healthy"],loc = 4)
    plt.title(f"{col} Stacked Bar Plot")
    plt.savefig(f"../figures/{col}_stacked_bar.png",bbox_inches="tight")
    plt.show()
```

```python
is_continuous = ["Age","RestingBP","Cholesterol","MaxHR", "Oldpeak"]
bins = [12 for _ in range(1,len(is_continuous))]
for col in is_continuous:
    df[col].plot.hist()
    plt.xlabel(col)
    plt.ylabel('count')
    plt.title(f"{col} Histogram")
    plt.savefig(f"../figures/{col}_hist.png",bbox_inches="tight")
    plt.show()
```

```python
for col in is_continuous:
    categories = df['HeartDisease'].unique()
    
    for c in categories:
        plt.hist(df[df['HeartDisease']==c][col],alpha=0.5,label=c,bins=20,density=True)
    plt.legend()
    plt.ylabel('fraction')
    plt.xlabel(col)
    plt.title(f"{col} Split Histogram")
    plt.savefig(f"../figures/{col}_split_hist.png",bbox_inches="tight")
    plt.show()
    
```
